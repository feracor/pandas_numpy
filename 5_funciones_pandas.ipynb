#groupby

```python
df_books = pd.read_csv('/content/bestsellers with categories.csv')
df_books.head(2)
```

```python
df_books.groupby('Author').count() #Agrupa los datos por una columna, convierte en el indice y cuenta
```

```python
df_books.groupby('Author').mean() #Agrupa los datos por una columna y muestra la media
```

```python
df_books.groupby('Author').min() #Agrupa los datos por una columna y muestra la mínima
```

```python
df_books.groupby('Author').sum() #Agrupa los datos por una columna y muestra la suma
```

```python
df_books.groupby('Author').sum().loc['William Davis']
```

```python
df_books.groupby('Author').sum().reset_index()
```

```python
df_books.groupby('Author').agg(['min','max']) #Agrupación por mínimo y máximo
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```

```python
df_books.groupby(['Author', 'Year']).count() #Agupación por llaves
```

#Combinando DataFrames

##concat

```python
dic= {'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3'],
      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3']}
```

```python
df1 = pd.DataFrame(dic)
df1
```

```python
dic2= {'A':['A4','A5','A6','A7'],
        'B':['B4','B5','B6','B7'],
        'C':['C4','C5','C6','C7'],
        'D':['D4','D5','D6','D7']}
df2 = pd.DataFrame(dic2)
df2
```

```python
pd.concat([df1,df2]) #Por defecto se concatenan los datos por filas. Los indices heredan el index original
```

```python
pd.concat([df1,df2], ignore_index=True) #De este modo se ignora el index original
```

```python
pd.concat([df1,df2], axis=1) #Forma de concatenar por columnas
```

##merge

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
      'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der) #Combina las tablas por columnas
```

```python
izq.merge(der, on='key') #Combina las tablas por columnas. La columna debe estar en ambas tablas
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
      'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2') #Se debe espeficiar el nombre de las columnas de cada dataframe por las cuales se desea unir
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
      'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2') #Para realizar la union, por defecto, solo trae regitros que coinciden (inner join)
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left') # Da prioridad a los datos del frame de la izq con la sentencia 'how='
```
