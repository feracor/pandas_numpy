#Funciones principales de Pandas

```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('/content/bestsellers with categories.csv')
df_books.head(2)
```

```python
df_books.info() #Entrega un resumen general del dataframe
```

```python
df_books.describe() #Entrega un resumen de la información estadística relevante
```

```python
df_books.tail(2) #Devuelve los últimos registros
```

```python
df_books.memory_usage(deep=True) #Informa cantidad de memoria que se usa
```

```python
df_books['Author'].value_counts() #Informa sobre la candidad de veces que se repitelos valores únicos en una columna específica
```

```python
df_books.iloc[0] #Para usar la función de 'drop', se genera un duplicado de una fila
```

```python
drop_dup = df_books.append(df_books.iloc[0]) #Se inserta la fila duplicada
drop_dup
```

```python
drop_dup.drop_duplicates() #Elimina filas duplicadas
```

```python
drop_dup.drop_duplicates(keep='last') #Elimina filas duplicadas, pero mantiene el último y elimina el último
```

```python
df_books.sort_values('Year') # Ordena los registros por columna específica
```

```python
df_books.sort_values('Year', ascending=False) # Ordena los registros por columna de forma descendiente
```
